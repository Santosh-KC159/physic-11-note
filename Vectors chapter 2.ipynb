### Chater:2  Vectors

###### Scalar Quantity

The physical quantities which are specified with the magnitude or size alone are scalar quantities. For example,  length, speed, work, mass, density, etc.

###### Vector Quantity

Vector quantities refer to the physical quantities characterized by the presence of both magnitude as well as direction. For example, displacement, force, torque, momentum, acceleration, velocity, etc.

##### Comparison between Scalars and Vectors

|Criteria|Scalar|Vector|
|--------|------|------|
|Definition|A scalar is a quantity with magnitude only.|A vector is a quantity with the magnitude as well as direction.|
|Direction|No direction|Yes there is the direction|
|Specified by|A number (Magnitude) and a Unit|A number (magnitude), direction and a unit.|
|Represented by|Quantity symbol|Quantity symbol in bold or an arrow sign above i.e.$\vec{a}$|
|Example|Mass and Temperature|Velocity and Acceleration|

###### Graphical Representation of Vectors

Vectors are drawn as arrows. An arrow has both a magnitude (how long it is) and a direction (the direction in which it points). The starting point of a vector is known as the tail and the end point is known as the head.



A vector $\vec{a}$ is represented by a directed line AB whose tail is A and head is B.Thus,we can write,

$$\vec{AB} =\vec{a}$$

The magnitude of $\vec{a}$ is written in modulus as,

$$|\vec{AB}|=|\vec{a}|$$ (read as mod of vector $\vec{a}$)

$$ or,AB=a$$


###  Kinds of a vector

There are many types of vectors,some of them are described below:

###### Zero 

A zero vector is a vector when the magnitude of the vector is zero and the starting point(tail) of the vector coincides with the terminal point(head). In other words, for a vector $\vec{AB}$ the co-ordinates of the point $A$ are same as that of the point $B$ then the vector is said to be a zero vector and is denoted by $0$.

This follows that the magnitude of the zero vector is zero and the direction of such a vector is indeterminate

###### Unit Vector

A vector which has a magnitude of unit length is called a unit vector. Suppose if $\vec{a}$ is a vector having a magnitude $a$ then the unit vector is denoted by $\hat{n}$ in the direction of the vector $\vec{a}$ and has the magnitude equal to 1.

$$ \therefore \hat{n}=\frac{\vec{a}}{|a|}$$

It must be carefully noted that any two unit vectors must not be considered as equal, because they might have the same magnitude, but the direction in which the vectors are taken might be different.

##### Position Vector

If O is taken as reference origin and A is an arbitrary point in space then the vector $\vec{OA}$ is called as the position vector of the point.

Position vector simply denotes the position or location of a point in the three-dimensional Cartesian system with respect to a reference origin.

###### Co-initial Vectors
The vectors which have the same starting point are called co-initial vectors.



The vectors $\vec{AB}$ and $\vec{AC}$ are called co-initial vectors as they have the same starting point.

###### Like and Unlike Vectors

The vectors having the same direction are known as like vectors. On the contrary, the vectors having the opposite direction with respect to each other are termed to be unlike vectors.

###### Co-planar Vectors

Three or more vectors lying in the same plane or parallel to the same plane are known as co-planar vectors.

###### Collinear Vectors

Vectors which lie along the same line or parallel lines are known to be collinear vectors. They are also known as parallel vectors.



##### Equal Vectors

Two or more vectors are said to be equal when their magnitude is equal and also their direction is the same.



The two vectors shown above, are equal vectors as they have both direction and magnitude equal.

###### Displacement Vector

If a point is displaced from position A to B then the displacement AB represents a vector $\vec{AB}$ which is known as the displacement vector.

###### Negative of a Vector

If two vectors are the same in magnitude but exactly opposite in direction then both the vectors are negative of each other. Assume there are two vectors $\vec{a}$ and $\vec{b}$, such that these vectors are exactly the same in magnitude but opposite in direction then these vectors can be given by

$$\vec{a} = – \vec{b}$$

### Triangle law of Vector Addition

Triangle law of vector addition states that "when two vectors are represented as two sides of the triangle with the order of magnitude and direction, then the third side of the triangle represents the magnitude and direction of the resultant vector".

###### Triangle Law of Vector Addition Derivation

Consider two vectors P and Q that are represented in the order of magnitude and direction by the sides OA and AB, respectively of the triangle OAB.To find magnitude,we draw perpendicular BC from B to meet the line OC at pont C when produced. Let $\vec{R}$ be the resultant of vectors $\vec{P}$ and $\vec{Q}$.Then,we can written as,

$$\vec{R}=\vec{P}+\vec{Q}$$




$Magnitude$

From right angled triangle $\triangle OCB$,

$$(OB)^{2}=(OC)^{2}+(BC)^{2}$$

$$ or, (OB)^{2}=(OA+AC)^{2}+(BC)^{2}$$

$$ \therefore (OB)^{2}=(OA)^{2}+(AC)^{2}+2 OA.AC+(BC)^{2} .....(i)$$

From figure,$OB=|\vec{R}|$,$OA=|\vec{P}|$ and$AB=|\vec{Q}|$

In triangle $\triangle ACB$ with $\theta$ as the angle between $\vec{P}$ and $\vec{Q}$,we can write

$$cos\theta=\frac{AC}{AB}$$

$$ or, AC=ABcos\theta=Qcos\theta$$

$$\therefore AC=Qcos\theta ..........(ii)$$

And

$$sin\theta=\frac{BC}{AB}$$

$$ or, BC=ABsin\theta=Qsin\theta$$

$$\therefore BC=Qsin\theta ........(iii)$$

Now,from equation (i),(ii) and (iii),it can be written as
(after substituting AC and BC in eq.1)

$$R^{2}=P^{2}+2PQcos\theta+Q^{2}cos^{2}\theta+Q^{2}sin^{2}\theta$$ 

$$ or, R^{2}=P^{2}+Q^{2}(cos^{2}\theta+sin^{2}\theta)+2PQcos\theta$$

$$or, R^{2}=P^{2}+2PQcos\theta+Q^{2} $$

$$\therefore R=\sqrt{P^{2}+2PQcos\theta+Q^{2}}...........(iv)$$

Above equation is the magnitude of the resultant vectors$\vec{P}$ and$\vec{Q}$.

$Direction$

To determine the direction of the resultant vector, let $\phi$ be the angle between the resultant vector $\vec{R}$ and vector $\vec{P}$.

From triangle $\triangle OBC$,

$$tan\phi=\frac{BC}{OC}=\frac{BC}{OA+AC}$$

$$tan\phi=\frac{Qsin\theta}{P+Qcos\theta}$$

$$\therefore \phi=tan^{−1}(\frac{Qsin\theta}{P+Qcos\theta}) ......(v)$$

Above equation is the direction of the resultant vector.

#### Parallelogram law of Vector Addition 

Parallelogram law of vector addition states that,"If two vectors are represented both in magnitude and direction by the two adjacent sides of parallelogram drawn from a point,then their resultant is represented in magnitude and direction by the diagonal of the parallelogram passing through that point."

##### Analytical Treatment of Parallelogram Law of Vector Addition



We suppose that,two vectors $\vec{A}$ and $\vec{B}$ inclined to each other at angle$\theta$ are represented in magbitude and direction by the sides OP and OS of the parallelogram OPQS as shown in fig.

According to the parallelogram law of vector addition, the diagonal OQ of the parallelogram OPQS represents resultant of $\vec{A}$ and $\vec{B}$ both magnitude and direction.

###### Magnitude

In $\triangle QNO$,

$$OQ^{2}=ON^{2}+QN^{2}$$

$$ =(OP+PN)^{2}+QN^{2}$$

$$\therefore OQ^{2}=OP^{2}+PN^{2}+2 OP.PN+QN^{2} .......(i)$$

From right angled triangle,$\triangle ONQ$,

$OQ=|\vec{R}|=R$,$OP=|\vec{A}|=A$ and $PQ=|\vec{B}|=B$

Also,from right angled triangle $\triangle QNP$

$$cos\theta=\frac{PN}{PQ}$$

$$=\frac{PN}{B}$$

$$\therefore PN=Bcos\theta ............(ii)$$

$$sin\theta= \frac{QN}{PQ}$$

$$=\frac{QN}{B}$$

$$\therefore QN= Bsin\theta  ..........(iii)$$

Now,from equation (i),(ii) and (iii),we can written as,

$$R^{2}=A^{2}+B^{2}cos^{2}\theta+2ABcos\theta+B^{2}sin^{2}\theta$$

$$=A^{2}+B^{2}(cos^{2}\theta+sin^{2}\theta)+2ABcos\theta$$

$$=A^{2}+B^{2}+2ABcos\theta$$

$$\therefore R=\sqrt{A^{2}+B^{2}+2ABcos\theta} ...........(iv)$$

This gives the magnitude of the resulatant vectors $\vec{A}$ and $\vec{B}$.

###### Direction

Consider the resulatant vector$\vec{R}$ makes an angle $\beta$ eith vector$\vec{A}$.

In right angled triangle $\triangle QNO$,

$$tan\beta =\frac{QN}{ON}=\frac{QN}{OP+PN}  ..............(v)$$

Since,$OP=A $, $PN=Bcos\theta$ and $QN=Bsin\theta$

So,euation(v) can be written as,

$$tan\beta =\frac{Bsin\theta}{A+Bcos\theta}$$

$$\therefore \beta =tan^{-1}(\frac{Bsin\theta}{A+Bcos\theta}) .........(vi)$$

This gives the direction of resultant $\vec{R}$ of vectors$\vec{A}$ and $\vec{B}$.


#### Special Cases

###### Case a:When $\theta=0$ i.e. when two vectors act along the same direction,we have,

$$R=\sqrt{A^{2}+B^{2}+2ABcos0}$$

$$R=\sqrt{A^{2}+B^{2}+2AB}$$

$$R=\sqrt{(A+B)^{2}}$$

$$\therefore R=A+B$$

and, $$ tan\beta=\frac{Bsin0}{A+Bcos0} =\frac{B(0)}{A+B(1)}=0$$

$$or,\beta=0^{0}$$

Thus,When two vectors act along the same direction $(\theta=0^{0})$,the resultant has magnitude equal to the sum of magnitude of two vectors and acts along the direction of $\vec{A}$ and $\vec{B}$.

###### Case b: When $\theta=90^{0}$ i.e. when two vectors act at right angles to each other.

Under this condition,we have,

$$R=\sqrt{A^{2}+B^{2}+2ABcos90^{0}}$$

$$\therefore R=\sqrt{A^{2}+B^{2}$$

and,$$tan\beta=\frac{Bsin90^{0}}{A+Bcos90^{0}} =\frac{B(1)}{A+B(0)}=\frac{B}{A}$$

The direction of $\vec{R}$,

$$\beta=tan^{-1}(\frac{B}{A})$$

###### Case c: When $\theta=180^{0}$ i.e.,when two vectors act along opposite directions.

$$R=\sqrt{A^{2}+B^{2}+2ABcos180^{0}}$$

$$R=\sqrt{A^{2}+B^{2}+2AB(-1)}$$

$$R=\sqrt{A^{2}+B^{2}-2AB}$$

$$R=\sqrt(A-B)^{2}$$

$$\therefore R=\pm(A-B)$$

So,R=(A-B) or (B-A)

and,$$tan\beta=\frac{Bsin180^{0}}{A+Bcos180^{0}} =\frac{B(0)}{A+B(-1)}=0$$

$$tan\beta =tan0^{0} or tan180^{0}$$

$$ or,\beta = 0^{0} or 180^{0}$$

Thus,when two vectors act along opposite direction$(\theta=180^{0}$,the resultant has magnitude equal to the difference of their magnitude and acts along the direction$\vec{A}$ (for $\beta=0^{0}$) or the direction of $\vec{B}$(for$ \beta=180^{0}$).





####  Polygon law of vector addition:Addition of more than two vectors

Polygon law of vector addition states that if a number of vectors can be represented in magnitude and direction by the sides of a polygon taken in the same order, then their resultant is represented in magnitude and direction by the closing side of the polygon taken in the opposite order.
Let us find the resultant of four vectors $\vec{A}$,$\vec{B}$,$\vec{C}$ and $\vec{D}$



In traingle OKL,the vectors $\vec{A}$ and$\vec{B}$ are represented by the sides $\vec{OK}$ and $\vec{OL}$ taken in the same order. Therefore, from the triangle law of vector addition,the closing side $\vec{OL}$ taken in the opposite order represents the resultant of vectors $\vec{OK}$ and $\vec{KL}$.

Thusv,$$\vec{OK} +\vec{KL}=\vec{OL}   ..........(i)$$

By applying the traingle law of vector addition to the traingle OLM, It shows that the side $\vec{OM}$ is the resultant of vectors$\vec{OL}$ and$\vec{LM}$ i.e., 

$$\vec{OL}+\vec{LM} =\vec{OM}$$

Using eq(i), we get,

$$\vec{OK} +\vec{KL}+\vec{LM} =\vec{OM} ..........(ii)$$

Similarly applying the triangle law of vector addition to the triangle OMN, we get,

$$\vec{OM}+\vec{MN}=\vec{ON}$$

$$\vec{OK} +\vec{KL}+\vec{LM}+\vec{MN}=\vec{ON} ..........(iii)$$

Now,$\vec{OK}=\vec{A}$,$\vec{KL}=\vec{B}$,$\vec{LM}=\vec{C}$ and $\vec{MN}=\vec{D}$ and representing $\vec{ON}=\vec{R}$,then eq(iii) can be written as

$$\vec{A}+\vec{B}+\vec{C}+\vec{D}=\vec{R}$$

Hence,the resultant is represented by the closing side $\vec{ON}$ of the polygon taken in opposite order which is the polygon law of vector addition.

#### Multiplication of Vectors

The vector quantities can be multiplied by two ways which are:

* Dot product or scalar product of vectors

* Cross product or vector product of vectors

###### Dot product or scalar product of vectors



* Here $(ABcos\theta)$ is a scalar so the dot product of vectors is always scalar.So,dot product is also called scalar product.

eg:(a) $Work(W)=\vec{F}.\vec{S}$  (b)$Power(P)=\vec{F}.\vec{v}$

* Dot product of two vectors obeys commutative law i.e.

$\vec{A}.\vec{B}=\vec{B}.\vec{A}$

* Scalar products of two vectors will be maximum when $cos\theta=max$ i.e. $\theta=0$
$(\vec{A}.\vec{B})_{max}=AB$

* Scalar products of two vectors will be minimum when $cos\theta =min$ i.e.$\theta=90$
$$(\vec{A}.\vec{B})_{min}=0$$

* The scalar product of a vector by itself is termed as self dot product and is given by

$\vec{A}.\vec{A}=AAcos0=A^{2}$

i.e.$A=\sqrt{\vec{A}.\vec{A}}$

* The angle between two ectors is given by

$\theta=cos^{-1}[\sqrt{\frac{\vec{A}.\vec{B}}{AB}}]$

* Some properties of unit vectors $\hat{i}$,$\hat{j}$ and $\hat{k}$ in case of dot product are mentioned below:

(a) The dot product of same unit vectors gives one.

$$ \hat{i}.\hat{i} =|\hat{i}||\hat{i}|cos0$$

$$=1*1*1=1$$

Similarly,$\hat{j}.\hat{j}=1$ and$\hat{k}.\hat{k}=1$

##### Cross product or vector product of two vectors

If $\vec{A}$ and $\vec{B}$ be two vectors and $\theta$ be the angle between them,cross product of vectors $\vec{A}$ and $\vec{B}$ is defined as,

$$\vec{A}*\vec{B}=|\vec{A}||\vec{B}|sin\theta \hat{n}$$

$$=ABsin\theta \hat{n}$$   where $\hat{n}$ is unit vector.

The result is a vector which is perpendicular to both $\vec{A}$ and $\vec{B}$.

###### Special cases

* when $\vec{A}$ and $\vec{B}$ are parallel i.e.,$\theta=0$,then

$$\vec{A}*\vec{B}=ABsin0 \hat{n}$$

$$0*\hat{n}=\vec{0}$$

* When $\vec{A}$ and $\vec{B}$ are perpendicular i.e.,$\theta=90}$,then

$$\vec{A}*\vec{B}=ABsin90\hat{n}$$

$$=AB\hat{n}$$

* The cross product of similar unit vectors is zero i.e.,

$$\hat{i}*\hat{i}=|\hat{i}||\hat{i}|sin0$$

$$=1*1*0\hat{n}=0\hat{n}=\vec{0}=0$$

Similarly,$$\hat{j}*\hat{j}=0$$ and $$\hat{k}*\hat{k}=0$$

* The cross product of two mutually perpendicular vectors taken in cyclic order gives another unit vector.i.e

$$ \hat{i}*\hat{j}=|\hat{i}||\hat{j}|sin90\hat{k}$$

$$=1*1*\hat{k}$$

Similarly,

$$\hat{i}*\hat{k}=\hat{j}$$ and $$\hat{j}*\hat{k}=\hat{i}$$

But,$$\hat{j}*\hat{i}=-\hat{k}$$

since,$$\hat{i}*\hat{j}\neq\hat{j}*\hat{i}$$

$$\therefore \hat{i}*\hat{j}=-\hat{j}*\hat{i}$$

* The cross-product of two vectors $\vec{A}$ and$\vec{B}$ in a same coordinate system is:

$$\vec{A}=A_{x}\hat{i}+A_{y}\hat{j}+A_{z}\hat{k}$$

And $$\vec{B}=B_{x}\hat{i}+B_{y}\hat{j}+B_{z}\hat{k}$$

Then $$\vec{A}*\vec{B}=(A_{x}\hat{i}+A_{y}\hat{j}+A_{z}\hat{k})*(B_{x}\hat{i}+B_{y}\hat{j}+B_{z}\hat{k})$$

$$=A_{x}B_{y}\hat{k}+A_{x}B_{z}({-\hat{j}})+A_{y}B_{x}({-\hat{k}})+A_{y}B_{z}\hat{i}+A_{z}B_{x}\hat{j}+A_{z}B_{y}({-\hat{i}})$$

$$=A_{x}B_{y}\hat{k}-A_{x}B_{z}{\hat{j}}-A_{y}B_{x}{\hat{k}}+A_{y}B_{z}\hat{i}+A_{z}B_{x}\hat{j}-A_{z}B_{y}{\hat{i}}$$

$$=(A_{y}B_{z}-A_{z}B_{y})\hat{i}+(A_{z}B_{x}-A_{x}B_{z})\hat{j})+(A_{x}B_{y}-A_{y}B_{x})\hat{k}$$
